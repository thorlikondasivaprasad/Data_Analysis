# Handle missing Data:fillna,dropna,interpolate

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\smoha\Documents\city.csv",parse_dates=['day'])  
df.set_index('day',inplace=True)
df
```

# fillna()

```python
df1=df.fillna('0')
df1
```

```python
df1=df.fillna({
    'temparature':0,
    'windspeed':0,
    'event':'no event'
})                            #for specific rows likewise
df1
```

```python
#to carryforward the upper to fill 'ffill'

df1=df.fillna(method='ffill')
df1
```

```python
df1
```

```python
df1=df.fillna(method='bfill')  #backward fill of NaN values
df1
```

```python
df
```

# interpolate()

```python
# Here without ffill,bfill we can make NaN values as Meaningful from original data
#using 'interpolate' method ,we can get meaningful insight from data
df2=df1.interpolate()
print()
print(df2)     #observe the data below no fill
```

```python
# by using 'time ' in above to get the data as sequentially ordered by date or to guess or estimate

print(df2,'\n') 
print("after using method=time:",'\n')
df3=df.interpolate(method='time')
print(df3)                           #observe the data
```

# concatenation

```python
import pandas as pd

df1=pd.DataFrame({
    'city':['Hyd','CSK','RCB'],
    'humidity':[80,67,78],
    'temparature':[34,32,35],
})
df1
```

```python
df2=pd.DataFrame({
    'city':['new york','Canada','chicago'],
    'humidity':[60,56,54],
    'temparature':[23,24,25],
})
df2
```

```python
df3=pd.concat([df1,df2],ignore_index=True)
df3
```

# Merge

```python
import pandas as pd

df12=pd.DataFrame({
    'city':['Hyd','CSK','RCB'],
    'humidity':[80,67,78],
})
df12
```

```python
df22=pd.DataFrame({
    'city':['new york','Canada','chicago'],
    'temparature':[23,24,25],
})
df22
```

```python
df33=pd.merge(df12,df22,on="city",how="right",indicator=True)  #how="inner","outer","left","right"
df33
```

```python
df33=pd.merge(df12,df22,on="city",how="outer",indicator=True)
df33
```
