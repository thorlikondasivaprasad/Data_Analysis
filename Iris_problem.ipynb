```python
iris1,iris2,iris3= np.split(de,3)
print("iris 1 :\n",iris1)
print("iris 2 :\n",iris2)
print("iris 3 :\n",iris3)
```

```python
header= np.array(['sepal length','sepal width','petal length','petal width'])
print(header)
```

```python
iris_min, iris_max, iris_avg, iris_std, iris_var = np.array([]), np.array([]), np.array([]) , np.array([]), np.array([])
for i in range(iris.shape[1]-1):
    iris_min = np.append(iris_min, de[:,i].min())
    iris_max = np.append(iris_max, de[:,i].max())
    iris_avg = np.append(iris_avg, de[:,i].mean())
    iris_std = np.append(iris_std, de[:,i].std())
    iris_var = np.append(iris_var, de[:,i].var())

print("Minimum values:")
for i in range(len(header)):
    print('\t',header[i], end=' ')
    print('%.2f'%iris_min[i])
print("Maximum values:")
for i in range(len(header)):
    print('\t',header[i], end=' ')
    print('%.2f'%iris_max[i])
print("Mean values:")
for i in range(len(header)):
    print('\t',header[i], end=' ')
    print('%.2f'%iris_avg[i])
print("Standard Deviation values:")
for i in range(len(header)):
    print('\t',header[i], end=' ')
    print('%.2f'%iris_std[i])
print("Variance values:")
for i in range(len(header)):
    print('\t',header[i], end=' ')
    print('%.2f'%iris_var[i])
```
